# Setting up sockets for data streaming

```python
#import needed packages

import socket
import requests
import json
```

```python
#set up url and headers for twitter API request

from requests.structures import CaseInsensitiveDict

url = "https://api.twitter.com/2/tweets/search/stream"

headers = CaseInsensitiveDict()
headers["Authorization"] = "Bearer AAAAAAAAAAAAAAAAAAAAANJZdQEAAAAAHwvIB30%2F8LgT6pV0xUbw16I%2Fd%2B0%3DZtIvQH5IsqM1qWPDPdUDHsvFX1aC3MV6wl4XeHJ1b1sXhgXl6t"
```

```python
#function to send the tweets to the localhost
def send_tweets_to_spark(resp, tcp_connection):
    for line in resp.iter_lines():
        try:
            #print tweets on screen and send to tcp port
            full_tweet = json.loads(line)
            tweet_text = full_tweet['data']['text']
            print(tweet_text.encode('utf-8') + b'\n')
            print ("------------------------------------------")
            tcp_connection.sendall(tweet_text.encode('utf-8') + b'\n')
        
        except:
            #print "BLAD" on screen
            print("BLAD")
            
            #added recursion to keep the streaming running despite errors
            send_tweets_to_spark(resp, tcp_connection) 
```

```python
#set up a socket connection
TCP_IP = "localhost"
TCP_PORT = 9008
conn = None
s = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
s.bind((TCP_IP, TCP_PORT))
s.listen(1)
print("Oczekuję na połączenie....")
conn, addr = s.accept()
print("Połączono. Pobieranie tweetów.")

#call twitter API
resp = requests.get(url, headers=headers, stream = True)

#send the tweets to the localhost
send_tweets_to_spark(resp, conn)
```
